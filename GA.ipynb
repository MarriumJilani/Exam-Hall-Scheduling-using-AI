```python
import random

POPULATION_SIZE = 100
TOURNAMENT_SIZE = 5
CROSSOVER_PROBABILITY = 0.8
MUTATION_PROBABILITY = 0.1
MAX_GENERATIONS = 100

COURSES = ["C1", "C2", "C3", "C4", "C5"]
EXAM_HALLS = ["H1", "H2"]
TIME_SLOTS = [("T1", 2), ("T2", 2), ("T3", 2)] #randomly deciding
MAX_HOURS_PER_DAY = 6
CONFLICTS = [("C1", "C2"), ("C1", "C4"), ("C2", "C5"), ("C3", "C4"), ("C4", "C5")]
```

```python
def initialize_population(population_size):
    population = []
    for i in range(population_size):
        solution = []
        for course in COURSES:
            exam_hall = random.choice(EXAM_HALLS)
            time_slot = random.choice(TIME_SLOTS)
            solution.append((course, time_slot, exam_hall))
        population.append(solution)
    return population

pop=initialize_population(POPULATION_SIZE)
i=0
for i, p in enumerate(pop):
  print(i, p)

```

```python
CONFLICTS = {('C1', 'C2'): 10, ('C1', 'C4'): 5, ('C2', 'C5'): 7, ('C3', 'C4'): 12, ('C4', 'C5'): 8}
CONFLICTING_PAIRS = [('C1', 'C2', 10), ('C1', 'C4', 5), ('C2', 'C5', 7), ('C3', 'C4', 12), ('C4', 'C5', 8)]

max_hrs = {'H1': MAX_HOURS_PER_DAY, 'H2': MAX_HOURS_PER_DAY}
# Check if exam hall and time slot exist
      #  if hall in EXAM_HALLS and time_slot in TIME_SLOTS:
       #     hall_usage[hall][time_slot] += duration

conflicting_pairs = [('C1', 'C2', 10), ('C1', 'C4', 5), ('C2', 'C5', 7),
                     ('C3', 'C4', 12), ('C4', 'C5', 8)]

max_hrs = {'H1': MAX_HOURS_PER_DAY, 'H2': MAX_HOURS_PER_DAY}


def fitness_function(solution):
    hall_penalty = 10
    conflict_penalty = 100
    
    conflicts = 0
    conflicting_students = 0

    # Initialize exam hall usage dict
    hall_usage = {}
    for hall in EXAM_HALLS:
      hall_usage[hall] = {}

    # Count conflicts and hall usage
    for i, exam1 in enumerate(solution):
      for j, exam2 in enumerate(solution):
        #print(i,j,exam1, exam2)
        if i != j and (exam1[0], exam2[0]) in CONFLICTS and exam1[1] == exam2[1]:
          if exam1[2]!=exam2[2] and exam1[1][0]==exam2[1][0]:
            conflicts += 1
            conflicting_students += CONFLICTS[(exam1[0], exam2[0])]
           # print('exam1, exam2')
            #print(exam1, exam2)
            
      #print(conflicting_students)

      hall = exam1[2]
      time_slot = exam1[1]

      if time_slot not in hall_usage[hall]:
        hall_usage[hall][time_slot] = exam1[1][1]
      else:
        hall_usage[hall][time_slot] += exam1[1][1]

    # Calculate total hall usage penalty
    hall_penalty_total = 0
    for hall in EXAM_HALLS:
      for time_slot in TIME_SLOTS:
        if time_slot in hall_usage[hall]:
          if hall_usage[hall][time_slot] > MAX_HOURS_PER_DAY:
            hall_penalty_total += hall_penalty * (hall_usage[hall][time_slot] - MAX_HOURS_PER_DAY)

    # Calculate total conflict penalty
    conflict_penalty_total = conflicting_students * conflict_penalty

    # Calculate fitness score
    fitness_score = hall_penalty_total + conflict_penalty_total

    # Return fitness score
    return fitness_score

ex=[('C1', ('T1', 2), 'H2'), ('C2', ('T1', 2), 'H1'), ('C3', ('T1', 2), 'H2'), ('C4', ('T1', 2), 'H1'), ('C5',('T1', 2), 'H1')]
fit=fitness_function(ex)
print(fit)
scores = [fitness_function(individual) for individual in pop]
for i, f in enumerate(scores):
  print(i, f)
 
```

```python
def tournament_selection(population, tournament_size):
    num_groups = len(population) // tournament_size
    #print(num_groups)
    if num_groups == 0:
        num_groups = 1
    groups = [population[i:i+tournament_size] for i in range(0, len(population), tournament_size)]

    selected = []
    for group in groups:
        best = min(group, key=fitness_function)
        #print(best)
        selected.append(best)
    return selected


selection=tournament_selection(pop,TOURNAMENT_SIZE)

for i, s in enumerate(selection):
  print(i, s)
```

```python
def single_point_crossover(parent1, parent2):
    # randomly select a crossover point
    crossover_point = random.randint(1, len(parent1) - 1)

    # exchange the tails of the two parents
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]

    return child1, child2
```

```python
def mutation(solution):
    # randomly select a gene
    gene_index = random.randint(0, len(solution) - 1)
    course, timeslot, hall = solution[gene_index]

    # randomly change its value to a random value
    new_timeslot = random.choice(TIME_SLOTS)
    new_hall = random.choice(EXAM_HALLS)

    solution[gene_index] = (course, new_timeslot, new_hall)
    return solution
```

```python
def select_survivors(offspring, population, population_size):
    combined_population = population + offspring
    sorted_population = sorted(combined_population, key=fitness_function)
    return sorted_population[:population_size]
```

```python
def genetic_algorithm(population_size, tournament_size, crossover_prob, mutation_prob, num_generations):
    population = pop
    for i in range(num_generations):
      #print(f'Generation {i+1}:')
      offspring = []
      for j in range(population_size):
        tournament_selections = selection
        parent1 = random.choice(tournament_selections)
        parent2 = random.choice(tournament_selections)
        if random.random() < crossover_prob:
          child1, child2 = single_point_crossover(parent1, parent2)
          if random.random() < mutation_prob:
            child1 = mutation(child1)
          if random.random() < mutation_prob:
            child2 = mutation(child2)
          offspring.extend([child1, child2])
        else:
          offspring.append(parent1)
          offspring.append(parent2)
      population = select_survivors(offspring, population,population_size)
      #for i, s in enumerate(population):
        #print(i, s)

    return population

final=genetic_algorithm(POPULATION_SIZE,TOURNAMENT_SIZE,CROSSOVER_PROBABILITY,MUTATION_PROBABILITY,MAX_GENERATIONS)
for i, s in enumerate(final):
  print(i, s)
#scores = [fitness_function(individual) for individual in final]
#for i, f in enumerate(scores):
#  print(i, f)

#this code is open to generic implementation as all parameters can be defined using input
#i have used parameters from the example just for the sake of displaying outputs 
```

One benefit of using a genetic algorithm for the scheduling problem is that genetic algorithms can handle problems with a large number of variables and constraints, which is typical for scheduling problems.
Additionally, genetic algorithms can deal with non-linear and non-differentiable objective functions, which can make them useful for complex problems. 
However, one disadvantage of genetic algorithms is that they can be computationally expensive, especially when dealing with large-scale problems. Furthermore, genetic algorithms may not always converge to an optimal solution, and the quality of the solution can depend on the choice of parameters and the quality of the initial population. 
As a result, other optimization techniques, such as linear programming or constraint programming, may be more appropriate for certain types of scheduling problems. These techniques have their strengths and weaknesses, and selecting the best approach will depend on the specific characteristics of the problem at hand.
